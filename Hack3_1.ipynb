```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv(r'Train_H3.csv')
data.head(20)
```

```python
data.describe()
```

```python
data.info()
```

```python
obj_cols = ['married','house_ownership','car_ownership','profession','city','state', 'risk_flag']
```

```python
for x in ['married','house_ownership','car_ownership','profession','city','state', 'risk_flag']:
    data[x].value_counts().plot(kind = 'bar')
    plt.show()
```

```python
data.state.value_counts()
```

```python
data.profession.unique()
```

```python
from sklearn import preprocessing

x = data.drop(columns = ['married','house_ownership','car_ownership','profession','city','state']) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled, columns = x.columns)
df.head()
```

```python
df.describe()
```

```python
X_data = pd.concat([data[obj_cols].drop(columns = ['city', 'risk_flag']), df.drop(columns = ['Id', 'risk_flag'])], axis = 1)
X_data.state.replace({'Uttar_Pradesh[5]':'Uttar_Pradesh'}, inplace = True)
X_data.profession.nunique()
```

```python
X_data = pd.get_dummies(X_data, drop_first = False)
X_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_data, data.risk_flag, test_size=0.2)
```

```python
model1 = RandomForestClassifier(n_estimators = 90)
model1.fit(X_train, Y_train)
```

```python
model2= DecisionTreeClassifier(criterion='entropy', random_state=0)
model2.fit(X_train,Y_train)
```

```python
from sklearn import metrics
```

```python
pred1 = model1.predict(X_test)
```

```python
pred2 = model2.predict(X_test)
```

```python
acc1, roc1 = metrics.accuracy_score(Y_test, pred1), metrics.roc_auc_score(Y_test, pred1)
acc2, roc2 = metrics.accuracy_score(Y_test, pred2), metrics.roc_auc_score(Y_test, pred2)
```

```python
acc1
```

```python
acc2
```

```python
roc1
```

```python
roc2
```

```python
for i in [80, 90, 100]:
    model3 = RandomForestClassifier(n_estimators=i)
    model3.fit(X_train, Y_train)
    
    model4 = DecisionTreeClassifier(criterion='entropy', random_state=0)
    model4.fit(X_train, Y_train)
    
    pred3 = model3.predict(X_test)
    pred4 = model4.predict(X_test)
    
    acc3, roc3 = metrics.accuracy_score(Y_test, pred3), metrics.roc_auc_score(Y_test, pred3)
    acc4, roc4 = metrics.accuracy_score(Y_test, pred4), metrics.roc_auc_score(Y_test, pred4)
    
    print('Forest: For %.2f estimators: \n Accuracy = %.4f \n roc_auc score = %.4f' %(i ,acc3, roc3))
    print('Tree: For %.2f estimators: \n Accuracy = %.4f \n roc_auc score = %.4f' %(i ,acc4 ,roc4))
```
